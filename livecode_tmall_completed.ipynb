First, let's import the libraries we'll need

```python
import requests # request library that makes a request for us
from bs4 import BeautifulSoup # beatifulSoup library allows us to use methods to extract data
```

Now let's set the URL for the page we're scraping, make a request to get the HTML and parse it. 

```python
url = "https://list.tmall.com/search_product.htm?q=%B5%E7%C4%D4%B1%CA%BC%C7%B1%BE" # go to the browser and show the link
response = requests.get(url) # tell what variable is - box
html = response.content
scraped = BeautifulSoup(html, 'html.parser')
```

```python
print(scraped) # Explain what print does
```

```python
print(scraped.prettify()) # Eplaint what prettify does
```

## Sixth stop (10 minutes)


```python
first_image = first_product.find("a", class_ = "productImg").img['src'][2:]
print(first_image)
```

```python
images_list = []
for product in products:
    tag = product.find("a", class_ = "productImg").img
    if tag.has_attr('data-ks-lazyload'):
        product_image = tag['data-ks-lazyload'][2:]
        images_list.append(product_image)
    else:
        product_image = tag['src'][2:]
        images_list.append(product_image)
print(images_list)
```
