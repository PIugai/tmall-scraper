First, let's import the libraries we'll need

```python
import requests
from bs4 import BeautifulSoup
```

Now let's set the URL for the page we're scraping, make a request to get the HTML and parse it. 

```python
url = "https://list.tmall.com/search_product.htm?q=%B1%CA%BC%C7%B1%BE%B5%E7%C4%D4"
response = requests.get(url)
html = response.content
scraped = BeautifulSoup(html, 'html.parser')
```

```python
products = scraped.select(".product")
print(products)
```

```python
first_product = products[0]
print(first_product.prettify())
```

```python
first_title = first_product.find('p', class_ = 'productTitle').a['title']
# first_title = first_product.p.findNext('p').a.text
print(first_title)
```

```python
first_price = first_product.p.em.text
print(first_price)
```

```python
first_link = first_product.find('p', class_='productTitle').a['href'][2:]
print(first_link)
```

```python
first_image = first_product.find("a", class_ = "productImg").img['src'][2:]
print(first_image)
```

```python
prices = [] # List

for product in products:
    product_price = product.p.em.text
#     product_price_float = float(product_price.lstrip('¥'))
#     prices.append(product_price_float) # Create a Dictionary and append to List
    prices.append(product_price)
    
print(prices)
```

```python
titles = [] # List

for product in products:
    product_title = product.find('p', class_ = 'productTitle').a['title']
    titles.append(product_title) # Create a Dictionary and append to List
    
print(titles)
```

```python
def has_src_but_no_data(tag):
    return tag.has_attr('data-ks-lazyload') and not tag.has_attr('src')

images_list = []
for product in products:
    if product.a.find(has_src_but_no_data):
        product_images = product.find("a", class_ = "productImg").img['data-ks-lazyload'][2:]
        images_list.append(product_images)
    else:
        product_images = product.find("a", class_ = "productImg").img['src'][2:]
        images_list.append(product_images)
print(images_list)
```

```python
shops = [] # List

for product in products:
    product_shop_name = product.find('div', class_ = 'productShop').text.split()
    shops.append(product_shop_name) # Create a Dictionary and append to List
    
print(shops)
```

```python
shop_links = [] # List

for product in products:
    product_shop_link = product.find('a', class_ = 'productShop-name')['href']
    shop_links.append(product_shop_link[2:]) # Create a Dictionary and append to List
    
print(shop_links)
```

```python
products_links = [] # List

for product in products:
    product_link = product.find("a", class_ = "productImg")['href'][2:]
    products_links.append(product_link[2:]) # Create a Dictionary and append to List
    
print(products_links)
```

```python
import pandas as pd
```

```python
raw_data = {'Title': titles, 
        'Image': images_list, 
        'Price': prices,
           'Shop': shops,
           'Product links': products_links,
           'Shop links': shop_links}
df = pd.DataFrame(raw_data, columns = ['Title', 'Image', 'Price', 'Shop', 'Product links', 'Shop links'])
df
```

```python
print(df.to_csv(index=False))
```
